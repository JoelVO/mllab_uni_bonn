# Task 4.1 and 4.2

```python
class TwoLayerNN:
   
    def __init__(self, d1,d2,phi2,phi3):
        
        self.W1 = np.random.uniform(-1,1,(d2,d1))
        self.b2 = np.random.uniform(-1,1,d2)
        self.W2 = np.random.uniform(-1,1,d2)
        self.b3 = np.random.uniform(-1,1)
        self.phi2 = phi2
        self.phi3 = phi3
        self.d1 = d1
        self.d2 = d2
        
    def feedForward(self,X,backprop = False):
        W1_ext = np.multiply(self.W1,np.ones((len(X),self.d2,self.d1)))
        net2 = np.sum(np.multiply(X,np.transpose(W1_ext,axes=[1,0,2])),axis=-1).T + np.multiply(self.b2,np.ones((len(X),self.d2)))
        o2 = self.phi2(net2)
        W2_ext = np.multiply(self.W2,np.ones((len(X),self.d2)))
        net3 = np.sum(np.multiply(W2_ext,o2),axis = 1) + self.b3
        o3 = self.phi3(net3)

        if backprop:
            return(o2,net2,o3)
        
        else:
            return(o3)
    
    def backprop(self,X,Y):
        
        o2,net2,o3 = self.feedForward(X,True)
        
        W1_ext = np.multiply(self.W1,np.ones((len(X),self.d2,self.d1)))
        W2_ext = np.multiply(self.W2,np.ones((len(X),self.d2)))
        
        delta2 = 2*(Y - o3)
        delta1 = np.multiply(W2_ext.T,delta2*np.ones_like(delta2)).T
        
        gradW2 = np.dot(np.expand_dims(o2,axis=-1),[delta2*np.ones_like(delta2)])
        gradW2 = gradW2[np.arange(0,len(X),1),:,np.arange(0,len(X),1)]
        gradB3 = delta2*np.ones_like(delta2)

        gradW1 = np.dot(np.expand_dims(X,axis=-1),np.expand_dims(np.multiply(delta1,np.array(net2>0,dtype='float')),axis=1))
        gradW1 = gradW1[np.arange(0,len(X),1),:,np.arange(0,len(X),1)]
        gradB2 = np.multiply(delta1,np.array(net2>0,dtype='float'))
        
        return(np.mean(gradW1,axis=0).T, np.mean(gradB2,axis=0).T, np.mean(gradW2,axis=0).T, np.mean(gradB3,axis=0).T)
        
    def sgd(self,X,Y,nu,K,S):
        
        history = [np.sqrt(np.sum(np.power(Y-self.feedForward(X),2)))/len(Y)]
        t = trange(S, desc='Bar desc', leave=True)
        
        for s in t:
            
            if s%50 == 0:
                t.set_description("Least-squares error  %f" % history[-1])
                t.refresh()
                
            index = np.random.choice(len(X),min(K,len(Y)))
            B = np.array(X)[index]
            
            gradW1,gradb2,gradW2,gradb3 = self.backprop(B,Y[index])
            
            self.W1 = self.W1 + nu*gradW1
            self.W2 = self.W2 + nu*gradW2
            self.b2 = self.b2 + nu*gradb2
            self.b3 = self.b3 + nu*gradb3
            
            history.append(np.sqrt(np.sum(np.power(Y-self.feedForward(X),2)))/len(Y))
                
        
        return(history)
            
            
            
            
```

# Task 4.3

```python
n = 250

theta1,theta2 = np.random.uniform(0,2*np.pi,n),np.random.uniform(0,2*np.pi,n)
r1,r2 = np.random.uniform(0,1,n),np.random.uniform(1,2,n)

X1 = np.multiply(r1,[np.cos(theta1),np.sin(theta1)]).T
X2 = np.multiply(r2,[np.cos(theta2),np.sin(theta2)]).T

X = np.concatenate((X1,X2),axis=0)
Y = np.concatenate((-np.ones(n),np.ones(n)))

phi2 = lambda x: np.maximum(x,0)
phi3 = lambda x: x
```

```python
H,models,nus = [],[],[0.5,0.1,0.01,0.001,0.0001,0.00001]

for nu in nus:
    print("nu: ", nu)
    model = TwoLayerNN(2,20,phi2,phi3)
    history = model.sgd(X,Y,nu,20,50000)
    H.append(history)
    models.append(model)
```

```python
for i,h in enumerate(H):
    plt.plot(h,label = "nu: " + str(nus[i]))
    
plt.xlabel("iteration")
plt.ylabel("error")
plt.legend()
plt.grid(False)
plt.title("Error vs. iterations for sgd")
plt.show()
```

```python
fig,axs = plt.subplots(ncols=3,nrows=2,figsize = (15,10))

for i in range(len(H)):
    xx2, yy2 = np.meshgrid(np.arange(-2.2, 2.2, .01),
                         np.arange(-2.2, 2.2, .01))

    mesh = np.c_[xx2.ravel(), yy2.ravel()]

    M = np.array(models[i].feedForward(mesh) < 0,dtype = "int")
    mesh = np.array(M).reshape(xx2.shape)

    axs[i//3,i%3].contourf(xx2, yy2, mesh, cmap=plt.cm.coolwarm, alpha=0.3)
    axs[i//3,i%3].scatter(X[...,0][:n],X[...,1][:n],s=20)
    axs[i//3,i%3].scatter(X[...,0][n:],X[...,1][n:],s=20)
    axs[i//3,i%3].grid(False)
    axs[i//3,i%3].set_title("nu = " + str(nus[i]))
    axs[i//3,i%3].set_xlabel("x")
    axs[i//3,i%3].set_ylabel("y")
    
plt.show()
```

The bigger the number of iteration is, the less the error gets. However it stabilizes very quickly. Therefore increasing the number of iterations from 50000 does not really bring a lot for error reduction.

```python
from tensorflow import keras
import tensorflow.keras.layers as layers
```

## a)

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

num_classes = 10
x_train = np.reshape(x_train,(len(x_train),28*28))
y_train = keras.utils.to_categorical(y_train, num_classes)

x_test = np.reshape(x_test,(len(x_test),28*28))
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
model = keras.models.Sequential()
model.add(layers.Dense(128,input_shape = (28*28,),activation= "relu"))
model.add(layers.Dense(120,activation="relu"))
model.add(layers.Dense(10,activation="softmax"))

model.compile(loss = "categorical_crossentropy",optimizer = "sgd",metrics = ["accuracy"])
```

```python
history = model.fit(x_train,y_train,batch_size = 128,epochs=20,validation_data=(x_test,y_test))
```

```python
fig,axs = plt.subplots(ncols = 2,nrows = 1,figsize = (10,5))

axs[0].plot(history.history['accuracy'])
axs[0].plot(history.history['val_accuracy'])
axs[0].set_title('model accuracy')
axs[0].set_ylabel('accuracy')
axs[0].set_xlabel('epoch')
axs[0].legend(['train', 'val'])
axs[0].grid(False)

axs[1].plot(history.history['loss'])
axs[1].plot(history.history['val_loss'])
axs[1].set_title('loss value')
axs[1].set_ylabel('loss')
axs[1].set_xlabel('epoch')
axs[1].legend(['train', 'val'])
axs[1].grid(False)

plt.show()
```

# Task 4.6

```python
import mllab.pca
import cv2
import tensorflow as tf
```

```python
#Loading the pedestrian dataset

X_train = np.concatenate([mllab.pca.load_pedestrian_images("train",True),mllab.pca.load_pedestrian_images("train",False)],axis=0)/255
X_test = np.concatenate([mllab.pca.load_pedestrian_images("test",True),mllab.pca.load_pedestrian_images("test",False)],axis=0)/255

Y_train = np.concatenate((np.ones(int(len(X_train)/2)),np.zeros(int(len(X_train)/2))))
Y_test = np.concatenate((np.ones(int(len(X_test)/2)),np.zeros(int(len(X_test)/2))))
```

```python
#Creating a sampler for the data augmentation

#We made a data augmentation due to the main problem the network was having was because of the size of the 
#training dataset

#We augment the data with vertical reflections, chaging brightness and saturation, bluring, mixing images and
#loosing vertical and horizontal parts of the image

kernel = np.ones((4,4),np.float32)/16

def sample_images(q):
    while True:
    
        if q:
            
            entry = np.random.choice(len(X_train)-1)
            img,y = X_train[entry],Y_train[entry]
        else:
            entry = np.random.choice(len(X_test)-1)
            img,y = X_test[entry],Y_test[entry]

        t = np.random.choice(7)

        if t == 0:
            a = img[...,::-1,:]
        if t == 1:
            if y == 1:
                r = np.random.uniform(0,0.25)
                if q:
                    j = np.random.choice(1000)

                    img2 = X_train[1000+j]
                else:
                    j = np.random.choice(500)

                    img2 = X_test[500+j]

                a = r*img2 + (1-r)*img
            else:
                r = np.random.uniform(0,0.25)

                if q:
                    j = np.random.choice(1000)

                    img2 = X_train[j]
                else:
                    j = np.random.choice(500)

                    img2 = X_test[j]
                    
                a = r*img2 + (1-r)*img

        if t == 2:
            alpha = np.random.uniform(1,2.5)
            beta = np.random.choice([1,-1])*min(100,np.random.poisson(40))
            a = cv2.convertScaleAbs(255*img, alpha=alpha, beta=beta)/255

        if t == 3:
            a = cv2.filter2D(img,-1,kernel)

        if t == 4:
            if y == 0:
                ti = max(1,np.random.choice(3))
                for j in range(ti):
                    x1 = np.random.choice(40)
                    w = min(10,np.random.choice(50-x1))
                    image = np.copy(img)
                    image[:,x1:x1+w,:] = np.random.uniform(0,1,(3))*np.ones((100,w,3))
            else:
                x1 = np.random.choice(50)
                image = np.copy(img)
                image[30+x1:,:,:] = np.random.uniform(0,1,(3))*np.ones((100-30-x1,50,3))

            a = image

        if t == 5:
            if y == 1:
                ti = max(1,np.random.choice(3))
                for j in range(ti):
                    x1 = np.random.choice(40)
                    w = min(10,np.random.choice(50-x1))
                    image = np.copy(img)
                    image[:,x1:x1+w,:] = np.random.uniform(0,1,(3))*np.ones((100,w,3))
            else:
                x1 = np.random.choice(50)
                image = np.copy(img)
                image[30+x1:,:,:] = np.random.uniform(0,1,(3))*np.ones((100-30-x1,50,3))

            a = image

        if t == 6:
            ti = max(1,np.random.choice(3))
            for j in range(ti):
                x1 = np.random.choice(40)
                w = min(10,np.random.choice(50-x1))
                image = np.copy(img)
                image[:,x1:x1+w,:] = np.random.uniform(0,1,(3))*np.ones((100,w,3))

            x1 = np.random.choice(50)
            image = np.copy(img)
            image[30+x1:,:,:] = np.random.uniform(0,1,(3))*np.ones((100-30-x1,50,3))

            a = image

        if t == 7:
            a = img

        yield a,y
        
    
```

```python
def get_data_generator(sampler):
    def get_data():
        while True:
            img,y = next(sampler)
            
            yield img,y
    return get_data
```

```python
#Creating the data augmented generators

dg_train = tf.data.Dataset.from_generator(get_data_generator(sample_images([True])),
    output_types=(tf.float32, tf.float32),
    output_shapes=((100, 50, 3), ()) )

dg_test = tf.data.Dataset.from_generator(get_data_generator(sample_images([False])),
    output_types=(tf.float32, tf.float32),
    output_shapes=((100, 50, 3), ()) )

gen_batch_train = dg_train.batch(50)
gen_batch_val = dg_test.batch(50)
```

```python
#Examples for the augmented data

xs, ys = next(gen_batch_train.as_numpy_iterator())
pedestrians = xs[np.where(ys == 1)[0]]
back = xs[np.where(ys != 1)[0]]

fig,axs = plt.subplots(nrows = 2,ncols = 10,figsize = (15,10))

for i in range(10):
    axs[0,i%10].imshow(pedestrians[i])
    axs[0,i%10].grid(False)
    
for i in range(10):
    axs[1,i%10].imshow(back[i])
    axs[1,i%10].grid(False)
    
plt.show()
```

```python
#Creating the model

model = keras.models.Sequential()
model.add(layers.Conv2D(16,input_shape=(100,50,3), kernel_size=(3,3),activation = "relu"))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(32, kernel_size=(3,3),activation = "relu"))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Dropout(0.5))
model.add(layers.Flatten())
model.add(layers.Dense(128,activation = "relu"))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1,activation = "sigmoid"))

model.compile(loss = "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])
#model.load_weights("./checkpoints/checkpoint") Used to load the trained network
model.optimizer.lr.assign(0.001)
```

```python
#Callbacks to save the weights that perform the best and to reduce learning rate in case of not 
#improving accuracy on the validation dataset

NET_NAME = "./checkpoints/checkpoint"

save_callback = tf.keras.callbacks.ModelCheckpoint(NET_NAME, 
                                                   verbose=True,
                                                   save_weights_only=True,
                                                   save_best_only=True
                                                  )

reduce_lr_callback = tf.keras.callbacks.ReduceLROnPlateau(factor=0.2, verbose=1)
```

```python
#Training the model

history = model.fit(gen_batch_train,
                    epochs=10,
                    steps_per_epoch=200,
                    validation_data=gen_batch_val,
                    validation_steps=50,
                    shuffle=True,
                    callbacks=[save_callback, reduce_lr_callback])
```

```python
#Plotting the accuracy of our training and validation augmented datasets 
#We got an accuracy bigger than 99% on the validation dataset

fig,axs = plt.subplots(ncols = 2,nrows = 1,figsize = (10,5))

axs[0].plot(history.history['accuracy'])
axs[0].plot(history.history['val_accuracy'])
axs[0].set_title('model accuracy')
axs[0].set_ylabel('accuracy')
axs[0].set_xlabel('epoch')
axs[0].legend(['train', 'val'])
axs[0].grid(False)

axs[1].plot(history.history['loss'])
axs[1].plot(history.history['val_loss'])
axs[1].set_title('loss value')
axs[1].set_ylabel('loss')
axs[1].set_xlabel('epoch')
axs[1].legend(['train', 'val'])
axs[1].grid(False)

plt.show()
```
